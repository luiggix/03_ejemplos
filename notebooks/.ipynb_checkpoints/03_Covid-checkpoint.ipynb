```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from math import ceil
from scipy.optimize import curve_fit
from scipy.interpolate import CubicSpline
```

```python
fuente = 'fuente: https://github.com/CSSEGISandData/COVID-19.git'
```

```python
covid = pd.read_csv("../data/time_series_covid19_confirmed_global.csv")
```

```python
covid.head()
```

```python
covid = covid.drop(columns= ['Province/State','Lat','Long'])
covid
```

```python
covid_grupos = covid.groupby(['Country/Region']).sum()
```

```python
paises = [{'n':'Mexico', 'c':'black'},
          {'n':'US', 'c':'C1'}, 
          {'n':'Spain', 'c': 'C2'}, 
          {'n':'Brazil', 'c': 'C3'}, 
          {'n': 'China', 'c': 'C4'}]

xtext = len(mexico)

pargraf = {'legend':False, 'figsize':(20,10), 'logy':True, 
           'color':'silver', 'lw':1.0, 'ls':'-', 'grid':False,
           'xticks':[i for i in range(1, xtext, 30)]}

for p in covid_grupos.index:
    covid_grupos.loc[p].plot(**pargraf)

for p in paises:
    pargraf['color'] = p['c']
    pargraf['lw'] = 2.0
    pargraf['rot'] = 45
    covid_grupos.loc[p['n']].plot(**pargraf)

plt.show()
```

```python
pargraf = {'legend':False, 'figsize':(20,10), 'logy':True, 
           'color':'silver', 'lw':1.0, 'ls':'-', 'grid':False,
           'xticks':[i for i in range(1, xtext, 30)]}

for p in covid_grupos.index:
    covid_grupos.loc[p].plot(**pargraf)

for p in paises:
    pargraf['color'] = p['c']
    pargraf['lw'] = 2.0
    pargraf['rot'] = 45
    covid_grupos.loc[p['n']].plot(**pargraf)
    plt.text(xtext, list(covid_grupos.loc[p['n']])[-1], p['n'], color=p['c'], fontsize=12)

plt.title(fuente, loc='right')
plt.show()
```

```python
skip = 100
mex_orig = np.array(mexico)
xmex_orig = np.arange(0, len(mex_orig))
mex_data = np.array(mexico[::skip])
xmex_data = xmex_orig[::skip]

print(len(mex_orig), mex_orig)
print(len(xmex_orig), xmex_orig)

print(len(mex_data), mex_data)
print(len(xmex_data), xmex_data)

plt.figure(figsize=(10,6))
plt.plot(xmex_orig, mex_orig)
plt.scatter(xmex_data, mex_data, marker='.', c='k')
plt.show()
```

```python
cs = CubicSpline(xmex_data, mex_data)
mex_inter = cs(xmex_orig)

plt.figure(figsize=(10,6))
plt.plot(xmex_orig, mex_orig, c='C0')
plt.plot(xmex_orig, mex_inter, c='C1')
plt.scatter(xmex_data, mex_data, marker='.', c='k')
plt.show()
```

```python
pargraf = {'legend':False, 'figsize':(20,10), 'logy':True, 
           'color':'silver', 'lw':1.0, 'ls':'-', 'grid':False,
           'xticks':[i for i in range(1, xtext, 30)]}

for p in covid_grupos.index:
    covid_grupos.loc[p].plot(**pargraf)

for p in paises:
    pargraf['color'] = p['c']
    pargraf['lw'] = 2.0
    pargraf['rot'] = 45
    covid_grupos.loc[p['n']].plot(**pargraf)
    plt.text(xtext, list(covid_grupos.loc[p['n']])[-1], p['n'], color=p['c'], fontsize=12)

plt.plot(xmex_orig, mex_inter, c='blue', lw=3.0, alpha=0.5)
plt.title(fuente, loc='right')
plt.show()
```

```python
skip = 50
limit= 300
mex_orig = np.array(mexico)
xmex_orig = np.arange(0, len(mex_orig))

mex_datar = np.array(mexico[:limit:skip])
xmex_datar = xmex_orig[:limit:skip]

print(len(mex_orig), mex_orig)
print(len(xmex_orig), xmex_orig)

print(len(mex_datar), mex_datar)
print(len(xmex_datar), xmex_datar)

def func(x, a, b, c):
    return a * np.exp(a * x + b) + c

popt, pcov = curve_fit(func, xmex_datar, mex_datar, method='dogbox', max_nfev=2000)#, p0=[50, 12])

xmex_data_fit = xmex_orig[:400]
mex_data_fit = func(xmex_data_fit, *popt)

plt.figure(figsize=(10,6))
plt.plot(xmex_data_fit, mex_data_fit)
plt.plot(xmex_orig, mex_orig)
plt.scatter(xmex_datar, mex_datar, marker='.', c='k')
plt.show()
```
